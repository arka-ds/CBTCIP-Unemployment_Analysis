```python
df.rename(columns = { 'Region' : 'States'}, inplace = True)
```

```python
df.columns
```

```python
average_unemployment_rate = df.groupby('States')['Estimated Unemployment Rate (%)'].mean()
print('average_unemployment_rate in india is :', average_unemployment_rate)
```

```python
#Name of the state with highest unemployment rate

name_of_state_highest_unemployment_rate = average_unemployment_rate.idxmax()
print('Name of the state with highest unemployment rate :', name_of_state_highest_unemployment_rate)
```

```python
#Highest Unemployment rate of that state
highest_unemployment_rate = average_unemployment_rate.max()
print('Highest Unemployment Rate :', highest_unemployment_rate)
```

```python
#Name of the state with lowest unemployment rate
name_of_state_lowest_unemployment_rate = average_unemployment_rate.idxmin()
print('Name of the state with lowest unemployment rate :', name_of_state_lowest_unemployment_rate)
```

```python
#Lowest Unemployment rate of that state
lowest_unemployment_rate = average_unemployment_rate.min()
print('Lowest Unemployment Rate :', lowest_unemployment_rate)
```

```python
# Visualization of unemployment over the time
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df)
plt.xticks(rotation=45)
plt.title('Unemployment Rate Over Time')
plt.show()

```
